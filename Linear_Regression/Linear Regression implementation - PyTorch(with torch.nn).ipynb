```python
import torch
import torch.nn as nn
import torch.nn.functional as F
```

```python
x_train = torch.FloatTensor([[1,1],[2,2],[3,3]])
y_train = torch.FloatTensor([[10],[20],[30]])
```

```python
# 모델링
class class_LinearRegression(nn.Module): #Linear Regression 정의
    def __init__(self): 
        super().__init__()
        self.linear = nn.Linear(2,1) # 입력 차원이 2이고 출력 차원이 1인 선형 레이어 정의
    def forward(self,x): # 순전파 연산
        return self.linear(x)

model_LinearRegression = class_LinearRegression() #class_LinearRegression로 부터 모델 객체 생성
optimizer = torch.optim.SGD(model_LinearRegression.parameters(),lr=0.01) #SGD optimizer를 사용하여 모델의 파라미터를 업데이트

for step in range(2000): 
    prediction = model_LinearRegression(x_train) # 예측값 저장
    cost = F.mse_loss(prediction,y_train) # 손실함수 정의
    optimizer.zero_grad()
    cost.backward() # 손실함수 미분
    optimizer.step() #위의 값을 사용해 gradient Update
    if (step + 1) % 10 == 0:
        print(f"Step: {step + 1}")
        print("Current Value of [w,b]:")
        for name, param in model_LinearRegression.named_parameters():
            if name == "linear.weight":
                print(f"w = {param.data[0]}")
            elif name == "linear.bias":
                print(f"b = {param.data[0]}")
        print("=" * 50)
        
# 모델 테스트
x_test = torch.FloatTensor([[4,4]])
model_test = model_LinearRegression(x_test)
print('='*100)
print("Model with [4] (expectation:40): ",model_test.detach().item())
print('='*100)
```
