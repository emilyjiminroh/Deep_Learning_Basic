```python
import numpy as np
import matplotlib.pyplot as plt  #시각화
# keras
# 딥러닝 모델을 구축하고 학습시키기 위한 고수준 신경망 API 
from keras.models import Sequential 
from keras.layers import Dense

from tensorflow.keras.callbacks import LambdaCallback

# 콜백 함수 정의 - 10번 학습 당 한 번씩 업데이트 된 W, b의 중간 값 출력
def print_weights(epoch, logs):
    if epoch % 10 == 0:
        weights = model.get_weights()
        W = weights[0][0][0]
        b = weights[1][0]
        print(f"Epoch: {epoch}, W: {W:.2f}, b: {b:.2f}")
```

```python
# Data
x_data = np.array([[1],[2],[3]])
y_data = np.array([[10],[20],[30]])

# 콜백 객체 생성
print_weights_callback = LambdaCallback(on_epoch_end=print_weights)

# Model, Cost, Train
model = Sequential() # 순전파 모델 생성(foward(self,x)와 비슷)
model.add(Dense(1,input_dim=1)) # 입력차원이 1인 Dense 레이어 추가
model.compile(loss='mse',optimizer='adam') # 손실함수와 optimizer 정의
model.fit(x_data,y_data,epochs=1000,verbose=0, callbacks=[print_weights_callback]) # 모델 학습 

# 결과값 출력
model.summary() # 모델의 구조와 파라미터 개수 등 출력
print(model.get_weights()) #학습이 완료된 W. b 출력

# 모델 테스트
print(model.predict(np.array([4])))

#plot
plt.scatter(x_data,y_data)
plt.plot(x_data,y_data)
plt.grid(True)
plt.show()
```
