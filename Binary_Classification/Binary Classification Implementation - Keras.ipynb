```python
import numpy as np
from keras.models import Sequential
from keras.layers import Dense

# Data
x_data = np.array([[1,2],[2,3],[3,1],[4,3],[5,3],[6,2]])
y_data = np.array([[0],[0],[0],[1],[1],[1]])
```

```python
# Model, Cost, Train
model = Sequential() #순전파 모델 생성
model.add(Dense(1,activation='sigmoid')) #입력차원이 1인 활성화함수 Dense Layer 추가
model.compile(loss='binary_crossentropy',optimizer='sgd',metrics=['accuracy']) #손실함수와 optimizer 정의
model.fit(x_data,y_data,epochs=1000,verbose=1) # 모델 학습

# 결과값 출력
model.summary() # 모델의 구조와 파라미터 개수 등 출력

#Inference                      
print(model.get_weights())
print(model.predict(x_data))
```
