```python
import torch
import numpy as np
x_train = torch.FloatTensor([[1,2],[2,3],[3,4],[4,4],[5,3],[6,2]])
y_train = torch.FloatTensor([[0],[0],[0],[1],[1],[1]])
W = torch.randn([2,1],requires_grad =True)
b = torch.randn([1], requires_grad= True)
optimizer = torch.optim.SGD([W,b],lr=0.01)
```

```python
# 모델링
def model_BinaryClassification(x): #model을 sigmoid에 적용한 값을 반환
    return torch.sigmoid(torch.matmul(x,W)+b)

for step in range(2000):
    prediction = model_BinaryClassification(x_train) # 예측값
    cost = torch.mean((-1)*y_train*torch.log(prediction)+(-1)*(1-y_train)*torch.log(1-prediction)) #손실함수
    optimizer.zero_grad() 
    cost.backward()# 미분해 gradient 값
    optimizer.step()# 업데이트
    
    if step % 100 == 0:
        print("Step:", step)
        print("Prediction:", prediction.detach().numpy())
        print("Updated W:", W.detach().numpy())
        print("Updated b:", b.detach().numpy())


# 모델 테스트
x_test = torch.FloatTensor([[6,1]])
model_test = model_BinaryClassification(x_test)
print("Model with [6,1] (expectation: 1) in sigmoid: ",model_test.detach().item())
model_test_binary = np.round(model_test>0.5).type(torch.float32)
print("Model with [6,1] (expectation: 1) in np.round: ",model_test_binary.detach().item())
```
