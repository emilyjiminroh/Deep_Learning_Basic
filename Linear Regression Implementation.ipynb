```python
import tensorflow as tf
import numpy as np
```

```python
x_data = np.array([[1,1],[2,2],[3,3]], dtype = np.float32) # 2X3
y_data = np.array([[10],[20],[30]],dtype=np.float32) # 3X1
W = tf.Variable(tf.random.normal([2,1])) # y=wX+b  W 2X1
b = tf.Variable(tf.random.normal([1])) # b 1X1
```

```python
def model_LinearRegression(x):
    return tf.matmul(x,W)+b

def cost_LinearRegression(model_x):
    return tf.reduce_mean(tf.square(model_x-y_data)) #MSE 

def train_optimization(x): #y: cost,x: [W,b]   
    with tf.GradientTape() as g: 
        model = model_LinearRegression(x)
        cost = cost_LinearRegression(model)
    gradients = g.gradient(cost,[W,b])
    tf.optimizers.SGD(0.01).apply_gradients(zip(gradients,[W,b])) # Stochastic Gradient Descent,  
    print("Current Value of [w,b] = [[%.2f,%.2f],%.2f]"%(W[0,0],W[1,0],b)) #    
    
for step in range(2001):
    train_optimization(x_data)

print('='*100)
x_test = np.array([[4,4]],dtype=np.float32)
model_test = model_LinearRegression(x_test)
print("model for [4,4]: ",model_test.numpy())
print('='*100)
```
